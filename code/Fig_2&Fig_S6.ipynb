```python
import pandas as pd
import os
import networkx as  nx
import matplotlib.pyplot as plt
from stargazer.stargazer import Stargazer, LineLocation
import statsmodels.api as sm
import random
import numpy as np
from scipy.stats import ks_2samp
from powerlaw import plot_pdf, Fit, pdf,ccdf
import math
from matplotlib import ticker
from matplotlib.ticker import FuncFormatter
from itertools import product
```

```python
#Fig_2
###把每一个时期新增加权边的比值与每个时期总共药物数量放在一起
df=pd.read_csv('D:Fig_2/total_weighted.csv',encoding='gbk')
time=list(df['time'])
c1=(68/255,4/255,90/255)
c2=(65/255,62/255,133/255)
c3=(48/255,104/255,141/255)
c4=(31/255,146/255,139/255)
c5=(53/255,183/255,119/255)
c6=(145/255,213/255,66/255)
c7=(248/255,230/255,32/255)

#ratio
ratio_existing=list(df['ratio_existing'])
ratio_existing_new=list(df['ratio_existing_new'])
ratio_new_new=list(df['ratio_new_new'])
ratio_have_common_neighbor=list(df['ratio_have_common_neigbor'])
ratio_appear_with_common_neighbor=list(df['ratio_appear_with_common_neighbor'])
ratio_existing_existing=list(df['ratio_existing_existing'])
ratio_not_strong_tra=list(df['ratio_not_strong_tra'])
ratio_non_tra=list(df['ratio_non_tra'])
ratio_total_other=list(df['ratio_total_other'])


bottom1=[]
bottom2=[]
bottom3=[]
bottom4=[]
for i in range(len(ratio_existing)):
    bottom1.append(ratio_existing[i]+ratio_appear_with_common_neighbor[i])
    bottom2.append(ratio_existing[i]+ratio_have_common_neighbor[i])
    bottom3.append(ratio_existing[i]+ratio_existing_existing[i])
    bottom4.append(ratio_existing[i]+ratio_existing_existing[i]+ratio_existing_new[i])
fig = plt.figure(figsize=(8,5))
ax1=plt.subplot(111)

ax1.bar(time,ratio_existing,label='Type 1',color=c2,width=0.8,edgecolor='black',alpha=1)
#ax1.bar(time,ratio_have_common_neighbor,label='triadic closure',color=c4,width=0.5,
#        bottom=ratio_existing,alpha=1)
ax1.bar(time,ratio_appear_with_common_neighbor,label='Type 2',color=c4,width=0.8,edgecolor='black',
        bottom=ratio_existing,hatch='\\',alpha=1)
ax1.bar(time,ratio_not_strong_tra,label='Type 3',color=c4,hatch='/',width=0.8,edgecolor='black',
        bottom=bottom1,alpha=1)
#ax1.bar(time,ratio_total_other,label='other',width=0.5,bottom=bottom2,color=c5,alpha=1)
ax1.bar(time,ratio_non_tra,label='Type 4',
        width=0.8,edgecolor='black',bottom=bottom2,color=c6,hatch='\\',alpha=1)
ax1.bar(time,ratio_existing_new,label='Type 5',
        width=0.8,edgecolor='black',bottom=bottom3,color=c6,hatch='/',alpha=1)
ax1.bar(time,ratio_new_new,label='Type 6',
        width=0.8,edgecolor='black',bottom=bottom4,color=c6,hatch='|',alpha=1)


ax1.set_ylabel('ratio of different kinds of new edges',fontsize=15)
ax1.set_xlabel('time',fontsize=15)
ax1.set_xticklabels(time, rotation=90,fontsize=15)
ax1.set_ylim([-0.03,1.03])
ax1.set_yticklabels(['{:.1f}'.format(0.0),
                     '{:.1f}'.format(0.0),
                     '{:.1f}'.format(0.2),
                     '{:.1f}'.format(0.4),
                     '{:.1f}'.format(0.6),
                     '{:.1f}'.format(0.8),
                     '{:.1f}'.format(1.0)], fontsize=15)
#ax1.text(-3,1.05,'A',fontsize=20)
def to_percent(tmp,position):
    return '%1.0f'%(100*tmp)+'%'
plt.gca().yaxis.set_major_formatter(FuncFormatter(to_percent))
#font1 = { 'size': 13 }
#plt.legend(loc='lower center',bbox_to_anchor=(0.48, -0.58),ncol=3,prop=font1 )

ax2=ax1.twinx()
total_df=pd.read_csv('D:Fig_2/total_weighted_edges.csv',encoding='gbk')
total_weighted_edges=list(total_df['total_weighted_edges'])
ax2.plot(time,total_weighted_edges,color='red',marker='o',label='number of\ntotal edges')
ax2.set_ylabel('number of total edges',fontsize=15)
plt.yticks(fontsize=15)
ax2.ticklabel_format(style='sci',axis='y',scilimits=(0,0))
ax2.set_ylim([-60000,2560000])
handles1, labels1=ax1.get_legend_handles_labels()
handles2, labels2=ax2.get_legend_handles_labels()
font1 = { 'size': 15 }
plt.legend(handles1+handles2,labels1+labels2,loc='lower center',
           bbox_to_anchor=(0.5, -0.55),ncol=4,prop=font1 )
plt.savefig('D:Fig_2/Fig_2B.TIFF',dpi=300,bbox_inches='tight')
plt.show()
```
