```python
###Fig_3B
###weighted_NODF
df1=pd.read_csv('D:Fig_3/weighted/weighted_NODF.csv',encoding='gbk')
df2=pd.read_csv('D:Fig_3/weighted/null_random_NODF.csv',encoding='gbk')
df3=pd.read_csv('D:Fig_3/weighted/mean.csv',encoding='gbk')

time=list(df1['time'])
NODF_1=list(df1['NODF'])
NODF_2=list(df2['NODF'])
NODF_3=list(df3['mean'])

fig = plt.figure(figsize=(12,6))
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False


ax1=plt.subplot(1,1,1)

ax1.plot(time,NODF_1,c='green',marker='^',label='TCM')

ax1.plot(time,NODF_3,c='blue',marker='*',label='null model 1')
ax1.plot(time,NODF_2,c='red',marker='*',label='null model 2')


ax1.set_xlabel('time',fontsize=18)
ax1.set_ylabel('weighted NODF',fontsize=18)
ax1.set_xticklabels(time, rotation=90,fontsize=18)
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1))
ax1.axvline(0,ls="--",linewidth=0.3,c="black")
ax1.axvline(1.2,ls="--",linewidth=0.3,c="black")
ax1.axvline(4.81,ls="--",linewidth=0.3,c="black")
ax1.axvline(8.6,ls="--",linewidth=0.3,c="black")
ax1.axvline(12.68,ls="--",linewidth=0.3,c="black")
ax1.axvline(15.44,ls="--",linewidth=0.3,c="black")
ax1.axvline(18.11,ls="--",linewidth=0.3,c="black")
ax1.axvline(19,ls="--",linewidth=0.3,c="black")

ax1.text(-0.45,0.20,'before the\nend of Han\n Dynasty',fontsize=15)
ax1.text(2.1,0.19,'Wei,Jin,\nSouthern and\nNorthern\nDynasties'.center(3),fontsize=15)
ax1.text(6.1,0.20,'Sui,Tang\nand Five\nDynasties'.center(8),fontsize=15)
ax1.text(10.1,0.19,'Song,Jin\nand Yuan\nDynasties'.center(11),fontsize=15)
ax1.text(13.3,0.20,'Ming\nDynasty'.center(14),fontsize=15)
ax1.text(16.2,0.20,'Qing\nDynasty'.center(16),fontsize=15)
ax1.text(18.,0.20,'Modern\nTimes'.center(-5),fontsize=15)
ax1.set_ylim([-0.04,0.24])
ax1.set_yticklabels(['{:.2f}'.format(0),
                     '{:.2f}'.format(0),
                    '{:.2f}'.format(0.05),
                    '{:.2f}'.format(0.10),
                    '{:.2f}'.format(0.15),
                    '{:.2f}'.format(0.20),
                    ],fontsize=18)

font1 = { 'size': 15 }
plt.legend(loc='upper left',bbox_to_anchor=(0.02, 0.8),ncol=1,prop=font1 )


plt.savefig('D:Fig_2/weighted/compare_nestedness_total.TIFF',dpi=300,bbox_inches='tight')
plt.show()
```

```python
# Fig_S13
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams

plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.sans-serif'] = ['Arial']


df = pd.read_csv('D:Fig_3/weighted/weighted_box.csv',encoding='gbk')
original_df = pd.read_csv('D:Fig_3/weighted/weighted_original.csv',encoding='gbk')


print(df.head())
err_color=sns.color_palette()[1]


plt.figure(figsize=(10, 6))


sns.lineplot(x='time', y='NODF', data=df, errorbar='sd', 
            #  color=sns.color_palette()[0],
             color='gray',
             linestyle=':',
             err_style='bars',
             err_kws={'elinewidth': 2.5, 'ecolor':err_color},
             label='weighted NODF (Null Model)') 


plt.plot(original_df['time'], original_df['original_NODF'], 
         color=sns.color_palette()[2],  
         linestyle='-',
         label='weighted NODF (Original data)')


plt.xticks(rotation=90)


plt.xlabel('time', fontsize=14)
plt.ylabel('weighted NODF', fontsize=14)

plt.legend()

plt.grid(False)


plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()

plt.savefig('D:Fig_3/weighted/original_degree_weighted_NODF.TIFF', dpi=300)

plt.show()
```
