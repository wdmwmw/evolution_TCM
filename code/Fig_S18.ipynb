```python
import pandas as pd
import os
import networkx as  nx
import matplotlib.pyplot as plt
```

```python
ls=['battery','vehicle','computer','medicine','nano','semiconductor','smartphone','solar','wind']
ls1=['battery','vehicle','computer','medicine','nano','semiconductor','smartphone','solar photovoltaic','wind']
colors=[(248/255,230/255,32/255),(68/255,4/255,90/255),'#FA7F6F','#63E398','#8ECFC6','#E7DAD2','#999999',
        (145/255,213/255,66/255),'#FFBE7A','#BEB8DC',
        '#274B83','#2C67A7','#3B85B6','#76B5AF','#9EC3AA','#C1D5BE',
       '#D6E4D2','#F2F3E9']
fig = plt.figure(figsize=(14,6))
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
ax1=plt.subplot(1,2,1)
#plt.plot(x, [0]*20, alpha=0)
#ax1.plot(x,stability,c=colors[0],label='TCM')
for u in range(len(ls)):
    df=pd.read_csv(f'D:Fig_S18/{ls[u]}/stability.csv',encoding='gbk')
    time=list(df['time'])
    total_core=list(df['total_core'])
    common_core=list(df['common_core'])
    ratio=list(df['ratio'])
    
    #x1=[str(i) for i in range(len(x)-len(time)+1, len(x)+1)]
    #print(x1)
    ax1.plot(time,ratio,linestyle='--',marker='*',c=colors[u+1],label=f'{ls1[u]}')
ax1.set_xlabel('time',fontsize=15)
ax1.set_ylabel('stability of Core',fontsize=15)
#font1 = { 'size': 13 }
#ax1.legend(loc='lower center',bbox_to_anchor=(0.5, -0.48),ncol=3,prop=font1 )
ax1.text(1910,1.03,'A',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

ax3 = ax1.twiny()
tcm_df=pd.read_csv('D:Fig_S18/original.csv',encoding='gbk')
tm=list(tcm_df['time'])
stability=list(tcm_df['stability'])
ax3.plot(tm,stability,c='red',marker='o',label='TCM')
ax3.set_xticklabels(tm, c='red',rotation=90,fontsize=13)
ax3.xaxis.set_ticks_position('top')
ax3.xaxis.set_label_position('top')

handles1,labels1=ax1.get_legend_handles_labels()
handles2,labels2=ax3.get_legend_handles_labels()
font1 = { 'size': 13 }
plt.legend(handles1+handles2,labels1+labels2,loc='lower center',bbox_to_anchor=(0.5, -0.43),ncol=2,prop=font1 )

ax2=plt.subplot(1,2,2)
box_df=pd.read_csv('D:Fig_S18/box.csv',encoding='gbk')
ls2=['TCM','battery','vehicle','computer','medicine','nano','semiconductor','smartphone','solar photovoltaic','wind']
ax2.boxplot([box_df.stability[box_df.technologies=='TCM'],box_df.stability[box_df.technologies=='battery'],
            box_df.stability[box_df.technologies=='vehicle'],box_df.stability[box_df.technologies=='computer'],
            box_df.stability[box_df.technologies=='medicine'],box_df.stability[box_df.technologies=='nano'],
            box_df.stability[box_df.technologies=='semiconductor'],box_df.stability[box_df.technologies=='smartphone'],
            box_df.stability[box_df.technologies=='solar photovoltaic'],box_df.stability[box_df.technologies=='wind']],
           labels=ls2,showmeans=True)

ax2.set_xlabel('technologies',fontsize=15)
ax2.set_ylabel('comparsion of first order statistics\nof different technologies',fontsize=15)
ax2.set_xticklabels(ls2, rotation=90,fontsize=13)
ax2.set_ylim([-0.05,1.05])
ax2.set_yticklabels(['{:.1f}'.format(0.0),
                     '{:.1f}'.format(0.0),
                     '{:.1f}'.format(0.2),
                     '{:.1f}'.format(0.4),
                     '{:.1f}'.format(0.6),
                     '{:.1f}'.format(0.8),
                    '{:.1f}'.format(1.0)], fontsize=13)
plt.axhline(y=0.5,xmin=0,xmax=7,linewidth=1.5,color='red')
font1 = { 'size': 13 }
#plt.legend(loc='lower center',bbox_to_anchor=(0.45, -0.51),ncol=3,prop=font1 )
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

ax2.text(-1.2,1.03,'B',fontsize=15)
#ax2.set_xticklabels(x, rotation=90,fontsize=15)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.23, hspace=0.2)
plt.savefig('D:Fig_S18/Fig_S18.TIFF',dpi=300,bbox_inches='tight')
plt.show()
```
